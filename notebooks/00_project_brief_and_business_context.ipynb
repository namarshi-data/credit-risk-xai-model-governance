# 00 - Project Brief and Business Context

## Canadian Retail Credit Risk Analytics: Explainable Default Prediction, Portfolio Monitoring & Model Governance

## Business problem

A Canadian retail lender needs an early-warning analytics solution to identify borrowers with elevated default risk, monitor portfolio quality, and support responsible credit-risk decisions.

The business problem is not only to predict whether a borrower may default. A practical credit risk solution must also explain the drivers of risk, quantify the trade-off between missed defaults and unnecessary manual reviews, and document controls that can be understood by credit strategy teams, portfolio managers, model validators, governance teams, and non-technical stakeholders.

This notebook defines the business framing, target outcome, stakeholder questions, modelling boundaries, explainability requirements, and governance expectations before any model is trained.

## Why this matters for finance roles in Canada

This project is designed to demonstrate practical skills commonly required in Canadian banking, credit risk, risk analytics, model validation, and data analytics roles.

It shows the ability to:

- Prepare and validate multi-sheet financial datasets using Python-style analytical workflows.
- Analyze borrower risk, loan exposure, default rates, missingness, and portfolio segments.
- Build baseline and challenger machine learning models for retail credit default prediction.
- Evaluate models using risk-relevant metrics rather than accuracy alone.
- Select an operating threshold based on business cost and review capacity.
- Use explainable AI techniques to communicate global and individual model drivers.
- Document model limitations, leakage controls, fairness considerations, and monitoring requirements.
- Translate technical model outputs into stakeholder-ready credit risk insights.

## Why Explainable AI Matters in Credit Risk

Explainable AI, also known as interpretable AI, refers to the ability of a model to provide understandable reasons for its predictions.

In credit risk, explainability is not optional. Lenders, risk teams, model validators, and business stakeholders need to understand why a borrower is classified as higher or lower risk.

Explainability matters in this project for six reasons:

#### 1. Regulatory and Governance Expectations

Financial institutions operate in highly controlled environments where credit decisions must be transparent, consistent, and auditable.

Explainable AI helps support governance by showing which borrower, loan, income, employment, and data-quality factors influenced the model’s prediction.

For this project, explainability is used to support model documentation, validation review, stakeholder communication, and responsible use of machine learning.

#### 2. Trust and Business Adoption

A model with strong predictive performance is not enough if credit risk teams cannot understand or challenge its outputs.

Explainable AI helps bridge the gap between technical modelling and business decision-making by showing the factors behind high-risk predictions.

This makes the model more useful for credit strategy, portfolio monitoring, manual review prioritization, and risk reporting.

#### 3. Bias Detection and Fairness Review

Machine learning models can unintentionally learn patterns from historical data that may create unfair or undesirable outcomes.

Explainability helps identify whether model predictions are being driven by reasonable risk factors or by sensitive or proxy-sensitive variables.

In this project, sensitive and proxy-sensitive fields are reviewed separately for governance purposes and excluded from the baseline modelling feature set.

#### 4. Model Validation and Auditability

Model validators and governance teams need to understand whether the model is stable, explainable, and aligned with the intended business use.

Explainable AI supports validation by showing the most important global risk drivers and by providing local explanations for individual predictions.

This project uses explainability outputs to support the model card, validation summary, stakeholder brief, and monitoring plan.

#### 5. Error Diagnosis and Model Improvement

When a model makes incorrect predictions, explainability helps identify possible causes.

For example, a false positive may be driven by missing borrower information, high interest rate, high loan-to-income ratio, or other risk signals.

These insights help analysts understand whether model errors are caused by data quality issues, weak features, unstable segments, or limitations in the model design.

#### 6. Human-AI Collaboration

Credit risk decisions often require human judgment.

Explainable AI allows analysts and stakeholders to combine domain knowledge with model-generated insights.

In this project, the model does not replace human decision-making. Instead, it supports analysts by ranking borrowers by risk, explaining the drivers of those predictions, and highlighting cases that may require additional review.

## Explainable AI Approach

This project uses machine learning models such as Logistic Regression, Random Forest, and XGBoost to estimate borrower default risk. The tree-based models are useful for identifying complex relationships between borrower characteristics, loan terms, income, employment profile, and data-quality indicators.

To make the model interpretable, the project uses an explainability toolkit consisting of:

`SHAP global explanations` to identify the most important drivers of default-risk predictions.

`SHAP dependence analysis` to understand how important features affect predicted risk across different value ranges.

`Local explanations` to show why individual borrowers receive higher or lower risk scores.

`Anchor-style rules` to translate model behaviour into business-readable decision patterns.

`Counterfactual scenarios` to test how changes in selected features may affect predicted risk.

Counterfactual scenarios are treated as diagnostic model-sensitivity analysis only. They are not customer instructions and should not be used as adverse-action explanations without additional validation, legal review, and governance approval.

## Initial Modelling Objective

The modelling objective is to estimate the probability that a borrower will default using borrower, loan, employment, income, and historical credit-behaviour attributes.

The model should help answer:

> Which borrowers or borrower segments show elevated default risk, and what factors are contributing to that risk?

The project compares interpretable baselines and tree-based machine learning models, including Logistic Regression, Random Forest, and XGBoost.

**Business interpretation:** A borrower flagged as high risk should be prioritized for additional review, enhanced verification, portfolio monitoring, collections strategy, or risk-management action.

In this project, the model is used as a decision-support tool, not as an automated credit-decline system.

## Success metrics

Because default is normally an imbalanced event, this project will not optimize for accuracy alone.

#### Primary model-selection metric: **PR-AUC**

PR-AUC is prioritized because it is more informative when the positive class is relatively rare and the business cares about correctly identifying higher-risk borrowers.

#### Secondary model metrics:

- ROC-AUC
- Recall for defaulters
- Precision among reviewed or high-risk borrowers
- F1 score
- Balanced accuracy
- Matthews correlation coefficient
- Brier score / probability calibration
- Confusion matrix analysis

#### Business metrics:

- Manual review rate
- Missed-default count
- False-positive review burden
- Business-cost curve
- Threshold comparison under operational review-capacity limits

## Governance boundaries

This project will explicitly review and document:

- Potential target leakage from repayment-derived variables.
- Sensitive or proxy-sensitive variables that should not be used in the baseline model.
- Explainability requirements for global and individual predictions.
- Threshold selection based on business cost and operational capacity, not an arbitrary 0.50 cutoff.
- Model monitoring requirements, including drift, stability, review rate, and performance decay.
- Limitations of counterfactual explanations, which are used for diagnostic analysis only and not as customer instructions.
- The difference between decision-support analytics and automated adverse credit decisioning.

The model should not be used for:
- Automated credit decline.
- Automated pricing decisions.
- Adverse-action communication.
- Production lending decisions without additional validation, legal review, fairness testing, monitoring, and approval from appropriate governance teams.
